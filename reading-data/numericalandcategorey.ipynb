# Barplot

```python
sns.barplot(x='Pclass', y='Age', hue='Sex', data=titanic) # x is categorical, y is continuous hue is used to differentiate the bars
# barplot gives the mean of the continuous variable
```

# histplot

```python
survived = titanic[titanic["Survived"] == 1]["Age"] # get the age of the people who survived
sns.histplot(survived, kde=False, bins=20) # kde is kernel density estimation, bins is used to set the number of bins
```

```python
sns.kdeplot(titanic[titanic["Survived"] == 1]["Age"]) # kde plot is used to plot the density of the continuous variable
sns.kdeplot(titanic[titanic["Survived"] == 0]["Age"]) # kde plot is used to plot the density of the continuous variable
sns.kdeplot(titanic["Age"]) # kde plot is used to plot the density of the continuous variable for the whole dataset
```
