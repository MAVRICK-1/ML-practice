```python
from sklearn.model_selection import train_test_split # Import train_test_split function meaning from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler # Import StandardScaler meaning from sklearn.preprocessing import StandardScaler
# StandardScaler is used to standardize the data means it is used to scale the data
```

```python
data = pd.read_csv('Social_Network_Ads.csv')
x_train, x_test , y_train, y_test = train_test_split(data[['Age', 'EstimatedSalary']], data['Purchased'], random_state = 0 , test_size=0.3) # Split the data into training and testing sets
# arguments are data[['Age', 'EstimatedSalary']], data['Purchased'] and random_state = 0 , random_state is used to fix the randomness of the data 0 means it is fixed, 
# fixed in the sense that the data will be same whenever we run the code, random_state = 1 means the data will be different whenever we run the code\
# test_size = 0.3 means 30% of the data will be used for testing and 70% of the data will be used for training
print(x_train.head())
print(y_train.head())
print(x_test.head())
print(y_test.head())
```

```python


scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(x_train)

# transform train and test sets
X_train_scaled = scaler.transform(x_train)
X_test_scaled = scaler.transform(x_test)
X_test_scaled, X_train_scaled
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns = x_train.columns)
```

```python
np.round(X_train_scaled.describe(), 1)
```
